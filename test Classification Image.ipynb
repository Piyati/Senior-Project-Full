```python
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

# โหลดโมเดลที่ฝึกสอนมาก่อนหน้านี้
model = load_model('/content/drive/MyDrive/Senior Project/Senior V 4 Grayscale/CLAHE/Output_InceptionResNetV2/CLAHE_image_InceptionResNetV2/InceptionResNetV2_CLAHE_best.h5')

# กำหนดขนาดภาพที่โมเดลรับข้อมูลเข้า
image_width = 256
image_height = 256

# โหลดและแปลงภาพเป็นรูปแบบที่ใช้ได้กับโมเดล CNN
image = load_img('/content/drive/MyDrive/Senior Project/Senior V 4 Grayscale/CLAHE/split_train_new_cell_png/ce/Snap_090.png_bl.png', target_size=(image_width, image_height))
image_array = img_to_array(image)
image_array = np.expand_dims(image_array, axis=0)

# ทำนายความน่าจะเป็นของแต่ละคลาสสำหรับภาพนี้
predictions = model.predict(image_array)

# แสดงผลลัพธ์
class_names = ['Cemento', 'Fibrous', 'Ossifying']  
for i, class_prob in enumerate(predictions[0]):
    class_name = class_names[i]
    probability = class_prob * 100
    print(f"Class: {class_name}, Probability: {probability:.2f}%")
plt.imshow(image)
```
